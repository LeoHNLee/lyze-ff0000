```python
import pandas as pd
import numpy as np
from datetime import datetime

import matplotlib.pyplot as plt
import seaborn as sns

import warnings, os, sys, re
warnings.filterwarnings('ignore')
```

```python
def upbitDate(x):
    month = int(x//1)
    day = round((x-month)*100)
    if month == 1 : year = 2019
    else : year = 2018
    return datetime(year, month, day)

def zestDate(x):
    year = int(x[:4])
    month = int(x[5:7])
    day = int(x[-2:])
    return datetime(year, month, day)
```

## MCO

```python
mcoCap = pd.read_csv('./dataset/origin/MCO/MCO_price.txt', sep='\t')
mco = pd.read_csv('./dataset/afterListing/MCO.csv')

mco['date'] = mco['date'].apply(upbitDate)
mco['volume']=np.log(mco['volume'])

mcoCap['DateTime'] = pd.to_datetime(mcoCap['DateTime'])
mcoCap['Volume'] = np.log(mcoCap['Volume'])

fig = plt.figure(figsize=(17,8))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()
sns.lineplot('date', 'volume', ax=ax1, color='red', data = mco[mco['date']<=datetime(2018,11,30)])
sns.lineplot('DateTime', 'Volume', ax=ax2, color='blue', data=mcoCap[mcoCap['DateTime']>=datetime(2018, 4, 20)])
fig.legend(['national', 'global'])
plt.xticks(rotation='vertical')
```
